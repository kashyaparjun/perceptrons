```python
x = 0.2
w = 0.5
b = 1
y = 1
alpha = 0.9
```

```python
for i in range(1000):
    forward = w*x + b
    act = 1/(1+(2.718281828)**(-forward))
    print("y_pred", act)
    error = (1-act)**2
    print("error", error)
    d_error = 2*(y-act)*(-1)
    d_pred = act*(1-act)
    d_w = x
    w = w - alpha*d_error*d_pred*d_w
    print("weight", w)
```
